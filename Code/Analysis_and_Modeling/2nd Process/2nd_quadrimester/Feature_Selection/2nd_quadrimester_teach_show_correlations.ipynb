```python
from scipy import stats
```

H0 of chi2 = Two variables are independents

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

```python
def get_result(p_value, independent_feature, target_feature, related_columns):
    if p_value <= 0.05:
        print(independent_feature+ ' and ' + target_feature + ' are dependents (Reject H0)')
        return True
    else:
        print(independent_feature + ' and ' + target_feature + ' are independents (fail to reject H0)')
        return False
```

```python
related_columns=[]
```

```python
for col in categorical_cols:
    chi2, p, dof, ex = chi2_contingency(pd.crosstab(dv_analys_record_personal_access[col],dv_analys_record_personal_access['abandona'],margins= False))
    if get_result(p, col,'abandona', related_columns):
        related_columns.append(col)
```

```python
related_columns
```

```python
for col in numerical_cols:
    p = stats.spearmanr(dv_analys_record_personal_access[col], dv_analys_record_personal_access['abandona']).pvalue
    if get_result(p, col, 'abandona', related_columns):
        related_columns.append(col)
```
