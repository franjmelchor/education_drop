```python
titulaciones = analys_teaching_record_personal_access['cod_plan'].cat.categories
```

```python
def delete_out_analys_values(df, df_concat):
    for t in titulaciones:
        analys_teaching_record_personal_access_t = analys_teaching_record_personal_access[
            analys_teaching_record_personal_access['cod_plan'] == t]
        df_t = df[df['cod_plan'] == t]
        df_t = df_t[df_t['expediente'].isin(
            analys_teaching_record_personal_access_t['expediente'])]

        df_concat = pd.concat([df_concat, df_t], ignore_index=True)
        
    return df_concat
```

```python
def get_cum_p_data(p, curso:int, semestre:str):
    p_data = final_plan_subject_call[(final_plan_subject_call['cod_plan'] == p.cod_plan) &
                   (final_plan_subject_call['expediente'] == p.expediente)
                                ].sort_values(by=['curso_academico'])
    if len(p_data.index) > 0:
        courses = p_data['curso_academico'].unique()[0:curso]

        p_data = p_data[p_data['curso_academico'].isin(courses)]
        if len(p_data.index > 0):
            if(semestre == '1S'):
                try:
                    no_valid_data = p_data[(p_data['curso_academico']==courses[curso-1]) & 
                                   (p_data['semestre'] == '2S')].index
                    p_data = p_data[p_data.index.isin(no_valid_data) == False]
                except:
                    return pd.Series()
    return(p_data)
    
```

```python
def get_course_p_data(p, curso:int, df:pd.DataFrame):
    p_data = df[(
        df['cod_plan'] == p.cod_plan) 
        &(df['expediente'] == p.expediente)
    ].sort_values(by=['curso_academico'])
    try:
        course = p_data['curso_academico'].unique()[curso-1]
        p_data = p_data[p_data['curso_academico'] == course]
    except:
        return pd.Series()
    return(p_data)
```

```python
def get_cum_core_ratio(p, curso:int, semestre:str):
    p_data = get_cum_p_data(p, curso, semestre)
    if len(p_data.index) > 0:
        n_core_subjects  = len(p_data[p_data['tipologia_asignatura'] == 'T'].index)
        n_passed_core_subjects = len(p_data[(p_data['tipologia_asignatura'] == 'T') &
                                            (p_data['nota_num'] >= 5.0)].index)
        if n_core_subjects > 0:

            return (n_passed_core_subjects / n_core_subjects)
        else:
            return -1
    else: 
        return -1
```

```python
def get_cum_median(p, curso:int, semestre:str):
    import numpy as np
    p_data = get_cum_p_data(p, curso, semestre)
    if len(p_data.index) > 0:
        return np.nanmedian(p_data['nota_num'])
    else:
        return -1
    
```

```python
def get_semester_median(p, curso:int, semestre:str):
    import numpy as np
    p_data = get_course_p_data(p, curso, final_plan_subject_call)
    if len(p_data.index) > 0:
        p_data = p_data[p_data['semestre']==semestre]
        return np.nanmedian(p_data['nota_num'])
    else:
        return -1
```

```python
def get_cum_absent_ratio(p, curso:int, semestre:str):
    p_data = get_cum_p_data(p, curso, semestre)
    if len(p_data.index > 0):
        n_subjects = len(p_data.index)
        n_absent_subjects = len(p_data[p_data['des_nota']=='NO PRESENTADO'].index)

        return (n_absent_subjects/n_subjects)
    else:
        return -1
```

```python
def get_scolarship(p, curso:int):
    p_data = get_course_p_data(p, curso,final_scolarship_per_year)
    if len(p_data.index) > 0:
        return (p_data['becario'].values[0])
    else:
        return -1
```

```python
def get_cum_pass_ratio(p, curso:int, semestre:str):
    p_data = get_cum_p_data(p, curso, semestre)
    if len(p_data.index) > 0: 
        n_subjects  = len(p_data.index)
        n_passed_subjects = len(p_data[p_data['nota_num'] >= 5.0].index)
        if n_subjects > 0:
            return (n_passed_subjects / n_subjects)
        else:
            return -1
    else:
        return -1
```

```python
def get_cum_more_1st_call_ratio(p, curso:int, semestre:str):
    p_data = get_cum_p_data(p, curso, semestre)
    
    if len(p_data.index) > 0:
        n_more_1st_call_subjects = len(p_data[p_data['numero_convocatorias_agotadas']>1])
        n_subjects  = len(p_data.index)

        if n_subjects > 0:
            return (n_more_1st_call_subjects / n_subjects)
        else:
            return -1
    else:
        return -1
```
